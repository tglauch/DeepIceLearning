```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = np.arange(6, 11)

h = np.sqrt(1. - 0.5**2)

x = np.concatenate([np.arange(ni) - i * 0.5 for i, ni in enumerate(n)])
y = np.repeat(h * (1 + np.arange(len(n))), n)

x = np.concatenate([x, 0.5 + x[::-1]])
y = np.concatenate([y, 2 * (len(n) + 0.5) * h - y[::-1]])
r = np.empty(len(x), dtype=[("x", np.float), ("y", np.float)])
r["x"] = x
r["y"] = y
s = np.argsort(r, order=("y", "x"))
x = x[s]
y = y[s]
```

```python
fig, ax = plt.subplots()
p = ax.scatter(x, y, s=450, c=1 + np.arange(len(x)), cmap=plt.cm.Spectral)
plt.colorbar(p, ax=ax)
ax.set_aspect("equal", adjustable="datalim", anchor="SW")
ax.set_facecolor("#212F3D")
for i in xrange(len(x)):
    ax.text(x[i], y[i], i+1, color="darkgrey", va="center", ha="center")
```
